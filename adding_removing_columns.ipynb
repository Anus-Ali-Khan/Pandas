## Adding/Removing Columns

```python
import pandas as pd
```

```python
coffee = pd.read_csv('./warmup-data/coffee.csv') 
```

```python
# Adding price column
coffee['price'] = 4.99
```

```python
import numpy as np

# add new_price column with specific condition for one coffee_type
coffee['new_price'] = np.where(coffee['Coffee Type'] == 'Espresso', 3.99, 5.99)
```

```python
# remove the price column
coffee.drop(columns=['price'],inplace=True)
# we can also remove column by this
# coffee[['Day','Coffee Type','Units Sold', 'new_price']] 
```

```python
coffee
```

```python
# in pandas if we write coffee_new = coffee so it will also change the original dataset so we  have to copy the data like this
coffee_new = coffee.copy()
coffee_new['price'] = 4.99
```

```python
coffee_new
```

```python
coffee
```

```python
coffee['revenue'] = coffee['Units Sold'] * coffee['new_price']
```

```python
coffee = coffee.rename(columns={'new_price':'price'})
```

```python
bios = pd.read_csv('./data/bios.csv')
```

```python
bios_new = bios.copy()
```

```python
bios_new['first_name'] = bios_new['name'].str.split(' ').str[0]
```

```python
bios_new.query('first_name == "Keith"')
```

```python
# Converting the born_date dataype to datetime
bios_new['born_datetime'] = pd.to_datetime(bios_new['born_date'])
```

```python
bios_new['born_year'] = bios_new['born_datetime'].dt.year
bios_new[['name','born_year']]
```

```python
# convert our new data to csv
bios_new.to_csv('./data/bios_new.csv',index=False)
```

```python
# lambda function used to write custom functions
bios['height_category'] = bios['height_cm'].apply(lambda x: 'Short' if x < 165 else ('Average' if x < 185 else 'Tall'))
```

```python
bios.head()
```

```python
def categorize_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
        return 'Lightweight'
    elif row['height_cm'] < 185 or row['weight_kg'] <= 80:
        return 'Middleweight'
    else:
        return 'Heavyweight'

bios['Category'] = bios.apply(categorize_athlete,axis=1)
```
