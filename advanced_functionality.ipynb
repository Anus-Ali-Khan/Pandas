### Advanced Functionality

```python
import pandas as pd
import numpy as np
```

```python
coffee = pd.read_csv('./warmup-data/coffee.csv')
bios = pd.read_csv('./data/bios.csv')

coffee['price'] = np.where(coffee['Coffee Type'] == 'Espresse',3.99,5.99 )
coffee['revenue'] = coffee['Units Sold'] * coffee['price']
```

```python
# shift : start the column from rows provided as .shift(2) moves two rows downward and .shift(-2) move two rows upward
coffee['yesterday_revenue'] = coffee['revenue'].shift(2)
```

```python
bios['height_rank'] = bios['height_cm'].rank(ascending=False)
```

```python
bios.sort_values(['height_rank']).sample(10)[['name','height_rank']]
```

```python
coffee.head()
```

```python
coffee['pct_change'] = coffee['revenue'] / coffee['yesterday_revenue'] * 100
```

```python
# cumsum 
coffee['cummulative_revenue'] = coffee['revenue'].cumsum()
```

```python
coffee.head()
```

```python
latte = coffee[coffee['Coffee Type'] == 'Latte'].copy()

# Sum the last 3 rows of unts Sold
latte['3day'] = latte['Units Sold'].rolling(3).sum()
```
